# Do Python exclude Null and Nan values?

```python
# When calculating the average, does Pandas include or exclude Null values?
print (df['YearsCode'].max())
print(df['YearsCode'].min())
print(df['YearsCode'].mean())
```

```python
before=df['YearsCode'].mean()
new_df=df.dropna(subset=['YearsCode'])
after=new_df['YearsCode'].mean()
print(before, after)
```
